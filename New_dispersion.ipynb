### Reading only the "I", "Q" and "U":

```python
I_stokes, Q_stokes, U_stokes = planck_new[0], planck_new[1], planck_new[2]
#print(I_stokes[:10], Q_stokes[:10], U_stokes[:10])
```

### Calculating the "P" and "Phi"

```python
P = np.sqrt(Q_stokes**2 + U_stokes**2)/I_stokes
Phi = 0.5 *np.arctan2(U_stokes, Q_stokes) + np.pi/2.
```

### Writing the Stokes Parameters, P and Phi into an numpy array

```python
stokes = np.array([I_stokes, Q_stokes, U_stokes, P, Phi])
```

```python
shape(stokes)
```

### Creating a Pandas DataFrame:

```python
df = pd.DataFrame({'I':stokes[0,:], 'Q': stokes[1,:], 'U': stokes[2,:], 'P': stokes[3,:], 'Phi(rad)':stokes[4,:]})
```

### Calculate, phi, theta, lon and lat

```python
lat = np.zeros(len(I_stokes))
lon = np.zeros(len(I_stokes))
import healpy as hp
newnside = 2048
for n in range(len(I_stokes)):
    (theta,phi)=hp.pix2ang(newnside,n)
    lat[n] = theta #Healpix has theta origin at NGP
    lon[n] = phi
```
