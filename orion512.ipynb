```python
from astropy.table import Table
from astropy.io import fits
import healpy  as hp
import pandas as pd
from kapteyn import wcs
planck_new= hp.fitsfunc.read_map('/Volumes/NANO PRO/HFI_SkyMap_353_2048_R2.00_full.fits',\
                         field=(0,1,2), nest=False, hdu=1, h=True, verbose=False, memmap=True)
I_STOKES, Q_STOKES, U_STOKES = planck_new[0], planck_new[1], planck_new[2]
newnside = 512
I_new = hp.ud_grade(I_STOKES, newnside)
Q_new = hp.ud_grade(Q_STOKES, newnside)
U_new = hp.ud_grade(U_STOKES, newnside)
P = np.sqrt(Q_new**2 + U_new**2)/I_new
Phi = 0.5 *np.arctan2(U_new, Q_new) + np.pi/2.
lat = np.zeros(len(I_new))
lon = np.zeros(len(I_new))
for n in range(len(I_new)):
    (theta,phi)=hp.pix2ang(newnside,n)
    lat[n] = np.pi/2.-theta #Healpix has theta origin at NGP
    lon[n] = phi
df_new = pd.DataFrame({'I': I_new[:],
                       'Q': Q_new[:],
                       'U': U_new[:],
                       'P': P[:],
                       'Phi': Phi[:],
                       'lon': lon[:],
                       'lat': lat[:]
                       })
```

### Setting a the region of interest based on the longitude and lattitude

```python
maxlat = (-8.)*np.pi/180.
minlat = (-24.0)*np.pi/180.
minlon = (203.4)*np.pi/180.
maxlon = (220.6)*np.pi/180.
print(maxlat, minlat, maxlon, minlon)
```

```python
DF1 = df_new[((df_new.lat <= maxlat) & (df_new.lat >= minlat))]
DF = DF1[((DF1.lon <= maxlon) & (DF1.lon >=minlon))] 
shape(DF)
```

```python
DF_matrix = numpy.array(DF.as_matrix(columns = ('lon', 'lat')))
shape(DF_matrix)
```
